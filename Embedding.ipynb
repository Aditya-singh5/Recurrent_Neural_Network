```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent = ['the glass of milk',
       'the glass of juice',
       'the cup of tea',
       'I am a good boy',
       'I am a good developer',
       'understand the meaning of word',
       'your videos are good']
```

```python
voc_size=10000
```

```python
one_hot_repr = [one_hot(words,voc_size)for words in sent]
```

```python
one_hot_repr
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_len = 8
embedded_doc = pad_sequences(one_hot_repr,padding='pre',maxlen=sent_len)
```

```python
embedded_doc
```

```python
dim = 10
```

```python
model=Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_len))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_doc)
```

```python
model.predict(embedded_doc[[0][0]])
```
